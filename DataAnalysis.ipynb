```python
dataDir = "/home/vinay/Documents/Data/UnifiedLogger/All-Data/augusta/"

# files = os.listdir(dataDir)
accelFiles = glob.glob(dataDir+"AccelerometerLog_*")
gpsFiles = glob.glob(dataDir+"GpsLog_*")
camLogFiles = glob.glob(dataDir+"CameraLog_*")
gyroFiles = glob.glob(dataDir+"GyroscopeLog_*")
videoFiles = glob.glob(dataDir+"Video_*")

# files = videoFiles
# print len(files)
# files

```

```python
# Parse GPS Files:
gpsLog = gpsFiles[3]
cols = ["time", "lat", "lon", "alt", "speed", "bearing"]
usecols = [1, 3, 5, 7, 9, 11]
gps = pd.read_csv(gpsLog, header=None, usecols=usecols, names=cols, delimiter=" ", skiprows=1)

m = folium.Map(location=[gps["lat"].mean(), gps["lon"].mean()], zoom_start=14)
folium.PolyLine(zip(gps['lat'], gps['lon']), color="red", weight=2.5, opacity=1).add_to(m)
m

# N = 2
# with open(gpsLog) as myfile:
#     head = [next(myfile) for x in xrange(N)]
# print head
```

```python
# Parse Accelerometer Files:
accelLog = accelFiles[0]
cols = ["t", "type", "x", "y", "z", "x?", "y?", "z?"]
accel = pd.read_csv(accelLog, header=None, names=cols, delimiter=";", skiprows=0)
accel

plt.plot(accel["t"], accel["x"], label="x")
plt.plot(accel["t"], accel["y"], label="y")
plt.plot(accel["t"], accel["z"], label="z")
plt.legend()
```

```python
# Parse Gyroscope Files:
gyroLog = gyroFiles[0]
cols = ["t", "type", "x", "y", "z", "x?", "y?", "z?"]
gyro = pd.read_csv(gyroLog, header=None, names=cols, delimiter=";", skiprows=0)
gyro

plt.plot(gyro["t"], gyro["x"], label="x")
plt.plot(gyro["t"], gyro["y"], label="y")
plt.plot(gyro["t"], gyro["z"], label="z")
plt.legend()
```
